```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Users\ADMIN\Desktop\apps.csv")
df
```

```python
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
avg_rating=df['Rating'].mean()
df['Rating']=df['Rating'].fillna(avg_rating)
```

```python
df['Rating'].isnull().sum()
```

```python
df['Installs'].unique()[:10]
```

```python
df['Installs'] = df['Installs'].str.replace('+', '', regex=False)
df['Installs'] = df['Installs'].str.replace(',', '', regex=False)
df['Installs'] = df['Installs'].astype(int)
```

```python
df['Installs'].dtype
```

```python
#The Installs column was cleaned by removing special characters and converting it into a numeric format to enable accurate popularity analysis.
```

```python
df['Price'].unique()[:10]
```

```python
df['Price']=df['Price'].str.replace('$','',regex=False)
df['Price']=df['Price'].astype(float)
```

```python
df['Price'].dtype
```

```python
df['Size'].unique()[:10]
```

```python
df['Size'] = df['Size'].fillna(df['Size'].mean())
```

```python
df['Size'].isnull().sum()
```

```python
category_count=df['Category'].value_counts()
category_count
```

```python
#The category-wise distribution reveals that the Google Play Store is dominated by apps from categories such as Family, Game, and Tools
plt.figure(figsize=(11,5))
sns.barplot(x=category_count.index, y=category_count.values)
plt.xticks(rotation=90)
plt.title("Distribution of Apps Across Categories")
plt.xlabel("Category")
plt.ylabel("Number of Apps")
plt.show()
```

```python
df['Rating'].describe()
```

```python
#he rating distribution shows that most applications on the Google Play Store are rated between 4 and 4.5, indicating generally positive user satisfaction
plt.figure(figsize=(8,5))
plt.hist(df['Rating'], bins=25)
plt.title("Distribution of App Ratings on Google Play Store")
plt.xlabel("Rating")
plt.ylabel("Number of Apps")
plt.show()
```

```python
df['Installs'].describe()
```

```python
#The install distribution is highly skewed, with most applications having relatively low download counts, while only a small fraction of apps achieve very high installs
plt.figure(figsize=(8,5))
plt.hist(df['Installs'], bins=30)
plt.xscale('log')
plt.title("Distribution of App Installs on Google Play Store")
plt.xlabel("Number of Installs (log scale)")
plt.ylabel("Number of Apps")
plt.show()
```

```python
df['Type'].value_counts()
```

```python
#he majority of apps on the Google Play Store are free, while a smaller portion are paid.
plt.figure(figsize=(6,5))
sns.countplot(x='Type', data=df)
plt.title("Free vs Paid Apps on Google Play Store")
plt.xlabel("App Type")
plt.ylabel("Number of Apps")
plt.show()
```

```python
#Scatter plot analysis of app size vs rating indicates that most apps are small to medium in size, and there is no clear correlation between app size and user rating. Users appear to value app quality independently of size.
plt.figure(figsize=(8,5))
sns.scatterplot(x='Size', y='Rating', data=df)
plt.title("App Size vs Rating")
plt.xlabel("Size (MB)")
plt.ylabel("Rating")
plt.show()
```
