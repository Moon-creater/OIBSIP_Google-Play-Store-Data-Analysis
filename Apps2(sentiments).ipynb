```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Users\ADMIN\Desktop\user_reviews.csv")
df
```

```python
df.info()
df.isnull().sum()
```

```python
df = df.dropna(subset=['Translated_Review'])
```

```python
df.shape
```

```python
df = df.dropna(subset=['Sentiment'])
```

```python
#Missing review texts and sentiment labels were removed to ensure accurate sentiment analysis, resulting in a clean dataset of valid reviews.
```

```python
#SENTIMENT ANALYSIS
sentiment_counts = df['Sentiment'].value_counts()
sentiment_counts
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, color='coral')
plt.title("Sentiment Distribution of Google Play Store Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()
```
